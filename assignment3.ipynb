## Visualization Technique

In this dashboard, four different visualization techniques were combined to analyze Apple product pricing trends, category differences, customer ratings, and product popularity.
Each visualization provides a different perspective of the dataset. While individual charts can answer specific questions, combining them into a dashboard allows users to explore multiple aspects of the data simultaneously.

### Line Chart: Price Trends Over Time

The line chart shows changes in average Apple product prices over time, separated by product category. This visualization is useful for identifying trends and changes in continuous data. Line charts are most effective when analyzing patterns across time because the connected points emphasize increases, decreases, and overall movement.

For this dashboard, the line chart helps answer questions such as: prices changes from 2020 to 2026, if products follow similar pricing trends, and if any categories becoming more expensive over time.

### Bar Chart: Average Price by Product Category

The bar chart compares average current prices across Apple product categories.

Bar charts are effective when comparing separate groups because the differences between categories can be quickly identified. In this dashboard, the bar chart highlights the difference between higher-priced products, such as Macs, and lower-priced products, such as Apple Watches.

### Scatter Plot: Price, Rating, and Popularity

The scatter plot examines the relationship between Apple product pricing and customer ratings while incorporating review volume. Scatter plots are useful for identifying relationships between variables and detecting patterns that may not be visible in summary statistics.

The visualization uses: Current price range and average customer rating on the two axis. And uses color to show category with bubble size reflecting total number of reviews that point represents.  The scatter plot helps explore whether more expensive Apple products receive higher ratings or whether popularity differs between product categories.

### Box Plot: Price Distribution

The box plot displays the distribution of prices within each product category.

Unlike the bar chart, which only displays average values, the box plot shows the median, variation, and potential outliers. This provides a better understanding of how consistent pricing is within each category.

### Combining Visualizations in an Interactive Dashboard

The four visualizations complement each other:

- The bar plot complements the line chart because the line chart focuses on change over time, while the bar chart provides a simple summary comparison between categories.
- The box plot complements the other charts by showing the spread of prices rather than only comparing averages.
- The scatter plot diplays the spread of the data, which helps contextualize the other charts.

The dashboard also includes interactive filtering options. Users can select a product category, marketplace platform, product condition, or price range, and all visualizations update simultaneously. This allows users to investigate specific questions, with the ability to see how the changes impact each graph.

## Visualization Library

This dashboard was created using Plotly Express and ipywidgets.

### Plotly Express

Plotly is an open source visualization library designed for creating interactive charts. Plotly was selected because it provides built in features such as hovering, zooming, and interactive legends, making it suitable for dashboard development. Unlike static visualization libraries, Plotly allows users to directly explore visualizations rather than only viewing a final image.

### ipywidgets

ipywidgets is an open-source library that provides interactive controls for Jupyter notebooks. It was used to create dropdown menus and sliders that connect user selections with updated visualizations.
 The combination of Plotly and ipywidgets allows this project to function as an interactive dashboard while remaining within a notebook environment.

### Framework Approach and Limitations

Plotly uses a declarative approach where users specify what variables should be mapped to visual elements rather than manually drawing each component. This makes it easier to create complex interactive visualizations with less code. A limitation of using Plotly with ipywidgets is that deploying dashboards outside of notebook environments requires additional tools or hosting solutions.

ipywidgets integrates directly with Jupyter environments, allowing controls such as dropdown menus and sliders to interact with Python functions. This framework was selected because it provides enough functionality for an interactive dashboard an is still basic enough.

## Demonstration

### Dataset Selection

The data set I picked was from kaggle: https://www.kaggle.com/datasets/ashyou09/apple-products-pricing-dataset-2020-2026?resource=download

The file contains Apple product pricing information from September 2020 through July 2026. Each row represents an Apple product listing from either Amazon or Flipkart.

The dataset contains:

- about 80,000 product listings
- 15 variables
- Four product categories:
    - iPhone
    - iPad
    - Mac
    - Watch

Important variables include:

- Product category
- Current price
- Launch price
- Discount percentage
- Product condition
- Customer rating
- Number of reviews

### Data Cleaning and Preparation

The dataset required several preparation steps before visualization:

1. Loaded the CSV file into pandas.
2. Converted the Date column into datetime format.
3. Created a Month variable for time-based analysis.
4. Grouped data by product category for summary visualizations.
5. Created price ranges for the scatter plot to reduce visual clutter.
6. Removed missing values from aggregated results.

Because the dataset contained many individual product listings, aggregation was used to make patterns easier to interpret.

### Dashboard Development

The dashboard was developed in stages, one chart at a time. Each chart was designed to answer a specific question about Apple pricing. Next, the charts were combined into a single dashboard layout.

Lastly, interactive controls were added:

- Product Category dropdown
- Platform dropdown

- Price Range slider

These controls allow users to dynamically filter the dataset and observe how patterns change across different subsets.

### Troubleshooting (Issues I ran into while creating the Dashboard)

Interactive Widgets: When moving the dashboard between environments, some platforms did not fully support ipywidgets. Google Colab was selected because it provided better compatibility with interactive notebook widgets.

Multiple Chart Display: Initially, only some charts displayed when multiple Plotly figures were generated together. This occurred because notebook output handling can overwrite previous visualizations. The final dashboard used interactive_output from ipywidgets to maintain a persistent output area.

Scatter Plot Readability: The original scatter plot displayed too many individual points, making the visualization difficult to interpret. Products were grouped into price ranges and summarized by category to improve readability.
